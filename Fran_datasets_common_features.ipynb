```python
nf = pd.read_csv('features_nirmal_all_stopclickbait.csv')
br = pd.read_csv('features_barbara_all_stopclickbait.csv')
ex_nf = pd.read_csv('ForFran.csv')
ex_br = pd.read_csv('features_barbara_all_small.csv')
```

```python
all_features = pd.merge(nf, br,  on='id')

print(len(nf.keys()), len(br.keys()), len(ex_nf.keys()), len(ex_br.keys()))
```

```python
print(len(all_features.keys()), len(nf.keys()), len(br.keys()), len(ex_nf.keys()), len(ex_br.keys()))
```

```python
cols_to_find = all_features.columns.values
cols_to_lookin = ex_nf.columns.values
# cols_to_lookin
# cols_to_find
```

```python
cols_subset = list(set(cols_to_find).intersection(cols_to_lookin))
cols_subset
```

```python
dataset_challenge_training = ex_nf[cols_subset]
dataset_stopclick_val = all_features[cols_subset]
```
